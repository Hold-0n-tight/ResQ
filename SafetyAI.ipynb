```python
import torchvision.transforms as transforms

# 데이터 변환 정의
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # 이미지를 224x224 크기로 조정
    transforms.ToTensor(),  # 이미지를 Tensor로 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 이미지 정규화
])
```

```python
from torchvision import datasets

# 훈련 데이터셋 불러오기
train_dataset = datasets.ImageFolder(root='./data/NCT-CRC-HE-100K_sorted', transform=transform)

# 검증 데이터셋 불러오기
val_dataset = datasets.ImageFolder(root='./data/CRC-VAL-HE-7K_sorted', transform=transform)

# 데이터셋 크기 확인
print(f"훈련 데이터셋 크기: {len(train_dataset)}")
print(f"검증 데이터셋 크기: {len(val_dataset)}")
```

```python
from torch.utils.data import DataLoader

# 훈련 데이터셋 로더
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# 검증 데이터셋 로더
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# 데이터로더 크기 확인 (몇 배치가 있는지)
print(f"훈련 데이터로더 크기: {len(train_loader)}")
print(f"검증 데이터로더 크기: {len(val_loader)}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# CNN 모델 정의
class CNNModel(nn.Module):
    def __init__(self, num_classes=9):
        super(CNNModel, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        # 최종 fc1 입력 크기를 128 * 28 * 28로 수정
        self.fc1 = nn.Linear(128 * 28 * 28, 512)
        self.fc2 = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.pool(nn.ReLU()(self.conv1(x)))  # 첫 번째 컨볼루션 + 활성화 + 풀링
        x = self.pool(nn.ReLU()(self.conv2(x)))  # 두 번째 컨볼루션 + 활성화 + 풀링
        x = self.pool(nn.ReLU()(self.conv3(x)))  # 세 번째 컨볼루션 + 활성화 + 풀링
        x = x.view(-1, 128 * 28 * 28)  # 차원 변환 (flatten)
        x = nn.ReLU()(self.fc1(x))  # 완전 연결층 + 활성화
        x = self.fc2(x)  # 최종 출력
        return x

# 모델 초기화
model = CNNModel(num_classes=9)  # 클래스가 9개로 설정 (ADI, BACK, DEB, LYM, MUC, MUS, NORM, STR, TUM)

# 모델 구조 출력 (확인용)
print(model)
```

```python
# 손실 함수와 최적화 함수 설정
criterion = nn.CrossEntropyLoss()  # 다중 클래스 분류에 적합한 손실 함수
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Adam 최적화 사용
```

```python
# 훈련 루프
num_epochs = 7  # 훈련할 에폭 수

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # GPU가 있을 경우 GPU, 없으면 CPU 사용

# 모델을 선택한 디바이스로 이동
model = model.to(device)

for epoch in range(num_epochs):
    model.train()  # 모델을 훈련 모드로 설정
    running_loss = 0.0  # 에폭마다의 손실 합

    # 배치 단위로 훈련
    for i, (inputs, labels) in enumerate(train_loader):
        # 데이터를 선택한 디바이스로 이동
        inputs, labels = inputs.to(device), labels.to(device)

        # 이전 기울기 값 초기화
        optimizer.zero_grad()

        # 모델에 입력 데이터를 넣고 출력 받기
        outputs = model(inputs)

        # 손실 계산
        loss = criterion(outputs, labels)

        # 역전파
        loss.backward()

        # 가중치 업데이트
        optimizer.step()

        # 손실 값 누적
        running_loss += loss.item()

        # 100번째 배치마다 현재 진행 상황 출력
        if i % 100 == 99:
            print(f"[Epoch {epoch+1}, Batch {i+1}] loss: {running_loss / 100:.3f}")
            running_loss = 0.0

    # 에폭마다 손실 출력
    avg_loss = running_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")
```

```python
model.eval()  # 평가 모드로 전환
correct = 0
total = 0
test_loss = 0.0

with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        test_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

avg_test_loss = test_loss / len(val_loader)
accuracy = 100 * correct / total

print(f'Test Loss: {avg_test_loss:.4f}, Accuracy: {accuracy:.2f}%')
```
